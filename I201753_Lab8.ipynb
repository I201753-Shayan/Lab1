```python
import numpy as np
import pandas as pd
import math
```

```python
data = pd.read_csv("play_tennis.csv")
data = data.drop('day' , axis=1)
data
```

```python
def entropy_label(y):
    uniques , counts = np.unique(y,return_counts=True)
    total=0
    for i in counts:
        total = total + i / (len(y)) * math.log2(i/(len(y)))
        total=total * (-1) #negating sign
    return total
```

```python
def Entropy(x,y): #coloumn to predicting one
    uniques_x , counts_x = np.unique(x,return_counts=True)
    uniques_y , counts_y = np.unique(y,return_counts=True)
    entropy_lab = entropy_label(y)
    
    label_yes , label_no=[] , []
    for i_ux in uniques_x:
        count_y, count_n = 0,0
        for j in range(len(x)):
            if i_ux == x[j]:
                if data['play'][j]=='Yes':
                    count_y = count_y +1 
                else:
                    count_n = count_n +1
        label_yes.append(count_y)
        label_no.append(count_n)
        
    
    total_list=[]
    for i in range(0,len(uniques_x)): #starting from 0, for list indexes
        total = 0
        if (label_yes[i] == 0):
            total+= -(label_no[i]/counts_x[i] * math.log2( label_no[i]/counts_x[i]))    
        elif (label_no[i] == 0):
            total+= -(label_yes[i]/counts_x[i] * math.log2( label_yes[i]/counts_x[i]))    
        else:
            total+= -(label_no[i]/counts_x[i] * math.log2( label_no[i]/counts_x[i]))    
            total+= -(label_yes[i]/counts_x[i] * math.log2( label_yes[i]/counts_x[i]))    
        
        total_list.append(total)
        
        
    total_count = 0     
    for i in range(len(counts_x)):
        total_count= total_count + (total_list[i] * counts_x[i] / len(y))
    
    info_gain = entropy_lab - total_count
    print("Info gain: ",info_gain)
    return info_gain
        
```

```python
y = np.array(data['play'])
```

```python
info_gain_outlook= Entropy(np.array(data['outlook']) , y)
info_gain_temperature= Entropy(np.array(data['temp']),y)
info_gain_humidity= Entropy(np.array(data['humidity']),y)
info_gain_wind= Entropy(np.array(data['wind']),y)
```
